```python
page = requests.get('https://vignanam.org/index.htm#&panel1-1' , headers = headers)
```

```python
suop = BeautifulSoup(page.content, "html.parser")
```

```python
print(suop.prettify())
```

```python
table = suop.find('table', {'class':'deftable'})
```

```python
table
```

```python
language = []
for lang in table.findAll('a'):
    language.append(lang.get_text())
```

```python
language
```

```python
links = []
for lang in language:
    if lang == 'english':
        link = 'https://vignanam.org/index.htm#&panel1-1'
        links.append(link)
    else:
        link = f'https://vignanam.org/{lang}.htm#&panel1-1'
        links.append(link)
```

```python
print(links)
```

```python
from typing import Dict
import re
from collections import defaultdict
import time


def search_header(link : str):
    content_dict = defaultdict(list)
    sub_header_list = []
    counter = 0
    
    page = requests.get(link, headers = headers)
    d_soup = BeautifulSoup(page.content, 'html.parser')
    ul_link = d_soup.find('ul', {'class':'aqtree3clickable'})
    # print(ul_link.prettify())
    
    for subheader in ul_link.findAll('a', {'class':'link4'}):
            sub_header_list.append(subheader.get_text().replace('\xa0', ' '))
    
    
    for header in ul_link.findAll('a', {'class':'link1'}):
        main_header = header.get_text()
        # print(main_header)
        sloka_len = int(re.findall(r'\d+', main_header)[0])
        content_dict[' '.join([i for i in header.get_text().split()[:-1]])].append(sub_header_list[counter:counter + sloka_len])
        counter += sloka_len
    
    return content_dict
```

```python
content_dict = search_header(links[0])
# print(contect_dict.keys())

print(content_dict['Nitya Parayana Slokas'][0])
```
